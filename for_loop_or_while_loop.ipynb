```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python

for_times = []
# start timer here
for i in xrange(1000):
    foo = np.random.randn(1000)
    foo = list(foo)
    bar = []
    start = time.time()
    for num in foo:
        if num > 0:
            bar.append(num)
    end = time.time() - start
    for_times.append(end) 
# end timer


plt.plot(for_times)
mean = np.sum(for_times) / len(for_times)
print "mean time: %s" % mean
```

```python
# build list here
foo = np.random.randn(1000)
foo = list(foo)

while_times = []

for i in xrange(1000):
    foo = np.random.randn(1000)
    foo = list(foo)
    bar = []
    start = time.time()
    while foo:
        if foo[0] > 0:
            bar.append(foo.pop(0))
        else:
            foo.pop(0)
    end = time.time() - start
    while_times.append(end)
    
plt.plot(while_times)
mean = np.sum(while_times) / len(while_times)
print "mean time: %s" % mean
```
