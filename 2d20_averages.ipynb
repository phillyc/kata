In DnD 5th Edition, there is an "advantage" and "disadvantage" skill check.

When in "advantage", the player rolls 2 20 sided dice and keeps the higher of the two rolls.

When in "disadvantage", they roll 2 20 sided dice and keep the lower.

What's the average roll result in each case?

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# First, let's get a baseline of just one 20 sided die.
all_one_die_rolls = []
for i in xrange(1000):
    roll_one = np.random.randint(1,21)
    all_one_die_rolls.append(roll_one)

avg_1d20 = sum(all_one_die_rolls) / len(all_one_die_rolls)
print "Average 1d20 roll: %s" % avg_1d20
plt.plot(all_one_die_rolls)
plt.ylabel("1d20 result")
```

As we can see here, the results of a single die roll is all over the place. With the average result after 1000 rolls to be between 9-11.

```python
# Ok, now for the "advantage" case.
all_adv_rolls = []
for i in xrange(1000):
    roll_one = np.random.randint(1,21)
    roll_two = np.random.randint(1,21)
    if roll_one > roll_two:
        all_adv_rolls.append(roll_one)
    else:
        # If the second random is lower than or equal to the first, place in list.
        all_adv_rolls.append(roll_two)

avg_adv_2d20 = sum(all_adv_rolls) / len(all_adv_rolls)
print "Average advantaged 2d20 roll: %s" % avg_adv_2d20
plt.plot(all_adv_rolls)
plt.ylabel("2d20 adv result")
```

```python
# Ok, now for the "disadvantage" case.
all_dis_rolls = []
for i in xrange(1000):
    roll_one = np.random.randint(1,21)
    roll_two = np.random.randint(1,21)
    if roll_one < roll_two:
        all_dis_rolls.append(roll_one)
    else:
        # If the second random is higher than or equal to the first, place in list.
        all_dis_rolls.append(roll_two)

avg_dis_2d20 = sum(all_dis_rolls) / len(all_dis_rolls)
print "Average disadvantage 2d20 roll: %s" % avg_dis_2d20
plt.plot(all_dis_rolls)
plt.ylabel("2d20 dis result")
```

So we can see here, that introduction of the advantage or disadvantage of being able to roll 2 20 sided dice, and throw away the higher or lower score results in a bias of +3 or -3 respectively.

This of course is the "average" bias result, and you can see from the plots that it doesn't at all affect the variance of the rolls, so it's still quit possible to get "natural" 20s (even if it's less likely).
